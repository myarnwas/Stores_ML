## First Step : Check For Missing Values

```python
print("\nStep 1: Checking Missing Data")
missing_data = data.isnull().sum()
print(missing_data)
```

# Cleaning Data (If Found)

```python
if missing_data.sum() > 0:
    data.fillna(data.mean(), inplace=True)
    print("Missing Data Filled with Mean Values")
else:
    print("No Missing Data Found. The dataset is clean!")
```

# Second Step : Checking for Inconsistent Data

```python
print("\nStep 2: Checking for Inconsistent/Noisy Data")
print("Statistical Summary Before Cleaning:")
print(data.describe())
```

# Third Step : Visualizing Data (Histogram)

```python
print("\nStep 3: Visualizing Data")
columns = data.columns
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

for i, column in enumerate(columns):
    ax = axes[i // 3, i % 3]
    ax.hist(data[column], bins=20, color='skyblue', edgecolor='black')
    ax.set_title(f"Histogram of {column}")
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')
```
